```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
df = pd.read_csv('teste/PPF.txt')
```

```python
plt.plot(np.array(df['Wavelength (nm)']),np.array(df['Absorbance']))
plt.show()
```

```python
peak_indices, _ = find_peaks(df['Absorbance'], height=0.9, distance=50)
```

```python
peak_indices
```

```python
x = np.array(df['Wavelength (nm)'])
y = np.array(df['Absorbance'])
```

```python
plt.plot(x,y)
plt.plot(x[peak_indices], y[peak_indices], "x")
```
